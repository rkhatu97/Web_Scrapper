```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as bsoup
import pandas as pd
import numpy as np
```

```python
#my_url = "https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=947557db-dac5-411c-afde-649c66d8aebf&as-searchtext=sam"
products=[]
rating=[]
price=[]
containers=[]
for i in range(1,24):
    my_url = "https://www.flipkart.com/search?q=samsung+mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&as-searchtext=samsung+mobiles&sort=recency_desc&page="
    uClient = uReq(my_url+str(i))
    page_html = uClient.read()
    uClient.close()
    page_soup = bsoup(page_html, "html.parser")
#grabs each product
    containers.append(page_soup.findAll("div",{"data-id":True}))
for j in range(0,23):
    for a in containers[j]:
        if a.find('div', attrs={'class':'_3wU53n'}):
              products.append(a.find('div', attrs={'class':'_3wU53n'}).text)
        if a.find('div', attrs={'class':'hGSR34'}):
              rating.append(a.find('div', attrs={'class':'hGSR34'}).text)
        else:
              rating.append(0)
        if a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'}):
              price.append(a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'}).text[1:].replace(',',''))
        

        
```

```python
df = pd.DataFrame({'Product Name':products,'Price':price,'Rating':rating})
print(df)
```

```python
ram=[]
a_rom=[]
for j in range(0,23):
    for a in containers[j]:
        if a.find('ul', attrs={'class':'vFw0gD'}):
            a_rom.append(a.find('ul', attrs={'class':'vFw0gD'}).text.split('|'))
            ram.append(a.find('ul', attrs={'class':'vFw0gD'}).text.split('|')[0].split(' ')[0])
rom=[] 
for i in a_rom:
    if len(i)>2:
         rom.append(i[1].split('GB')[0].split(' ')[1])

print(ram)
print(rom)
```
